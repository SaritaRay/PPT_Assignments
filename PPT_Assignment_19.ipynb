```python
# SOLUTION: 1
# The time complexity of the algorithm is O(n * log(k)), where n is the total number of elements in all the linked lists and k is the number of linked lists. The space complexity of the algorithm is O(k).

# CODE:

import heapq

class listNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
        
class Solution:
    def mergeKLists(self, lists: list[listNode]) -> listNode:
        if not lists:
            return None
        if len(lists) == 1:
            return lists[0]
        mid = len(lists) // 2
        left = self.mergeKLists(lists[:mid])
        right = self.mergeKLists(lists[mid:])
        return self.merge(left, right)

    def merge(self, l1: listNode, l2: listNode) -> listNode:
        if not l1 or not l2:
            return l1 or l2
        if l1.val < l2.val:
            l1.next = self.merge(l1.next, l2)
            return l1
        else:
            l2.next = self.merge(l1, l2.next)
            return l2

# exmple usage:
# example 1:
input_lists = [[1,4,5],[1,3,4],[2,6]]
lists = []
for lst in input_lists:
    head = listNode(lst[0])
    node = head
    for i in range(1,len(lst)):
        node.next = listNode(lst[i])
        node = node.next
    lists.append(head)

s = Solution()
result = s.mergeKLists(lists)
output_list = []
while result != None:
    output_list.append(result.val)
    result = result.next

print("Output of Example_1 is: ", output_list)

print(" ")

# example 2:
input_lists = []
lists = []
for lst in input_lists:
    head = listNode(lst[0])
    node = head
    for i in range(1,len(lst)):
        node.next = listNode(lst[i])
        node = node.next
    lists.append(head)

s = Solution()
result = s.mergeKLists(lists)
output_list = []
while result != None:
    output_list.append(result.val)
    result = result.next

print("Output of Example_2 is: ", output_list)

print(" ")

# example 3:
input_lists = [[]]
lists = []
for lst in input_lists:
    head = listNode(lst)
    node = head
    for i in range(1,len(lst)):
        node.next = listNode(lst[i])
        node = node.next
    lists.append(head)

s = Solution()
result = s.mergeKLists(lists)
output_list = []
while result != None:
    output_list.append(result.val)
    result = result.next

print("Output of Example_3 is: ", output_list)
```

```python
# SOLUTION: 2
# The time complexity of this algorithm is O(n * log(n)), where n is the length of the input array. The space complexity of this algorithm is O(n).

# CODE:

class TreeNode:
    def __init__(self, val=0, left=None, right=None, count=0):
        self.val = val
        self.left = left
        self.right = right
        self.count = count

class Solution:
    def insert(self, root: TreeNode, val: int) -> tuple[TreeNode, int]:
        if not root:
            return TreeNode(val), 0

        if val < root.val:
            root.count += 1
            root.left, count = self.insert(root.left, val)
            return root, count

        else:
            root.right, count = self.insert(root.right, val)
            return root, root.count + count + (val > root.val)

    def countSmaller(self, nums: list[int]) -> list[int]:
        res = []
        root = None

        for num in reversed(nums):
            root, count = self.insert(root, num)
            res.append(count)

        return res[::-1]

# example usage:
# example 1:
nums = [5,2,6,1]
s = Solution()
print("Output of Example_1 is: ", s.countSmaller(nums))

print(" ")

# example 2:
nums = [-1]
s = Solution()
print("Output of Example_2 is: ", s.countSmaller(nums))

print(" ")

# example 3:
nums = [-1,-1]
s = Solution()
print("Output of Example_3 is: ", s.countSmaller(nums))
```

```python
# SOLUTION: 3
# The time complexity of this algorithm is O(n log n) and space complexity is O(n).

# CODE:

def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left = arr[:mid]
        right = arr[mid:]

        merge_sort(left)
        merge_sort(right)

        i = j = k = 0

        while i < len(left) and j < len(right):
            if left[i] < right[j]:
                arr[k] = left[i]
                i += 1
            else:
                arr[k] = right[j]
                j += 1
            k += 1

        while i < len(left):
            arr[k] = left[i]
            i += 1
            k += 1

        while j < len(right):
            arr[k] = right[j]
            j += 1
            k += 1

    return arr

# example usage:
# example 1:
nums = [5,2,3,1]
print("Output of example_1 is: ", merge_sort(nums))

print(" ")

# example 2:
nums = [5,1,1,2,0,0]
print("Output of example_2 is: ", merge_sort(nums))
```

```python
# SOLUTION: 4
# The time complexity of this algorithm is O(n) and the space complexity is O(1).

# CODE:

def pushZerosToEnd(arr, n):
    count = 0

    for i in range(n):
        if arr[i] != 0:
            arr[count] = arr[i]
            count += 1

    while count < n:
        arr[count] = 0
        count += 1

    return arr

# example usage:
# example 1:
arr = [1, 2, 0, 4, 3, 0, 5, 0]
n = len(arr)
print("Output of example_1 is: ", pushZerosToEnd(arr, n))

print(" ")

# example 2:
arr = [1, 2, 0, 0, 0, 3, 6]
n = len(arr)
print("Output of example_2 is: ", pushZerosToEnd(arr, n))
```

```python
# SOLUTION: 5
# The time complexity of the algorithm is O(n) and the space complexity is O(1).

# CODE:


def rightRotate(arr, n, outOfPlace, cur):
    temp = arr[cur]
    for i in range(cur, outOfPlace, -1):
        arr[i] = arr[i - 1]
    arr[outOfPlace] = temp
    return arr


def rearrange(arr, n):
    outOfPlace = -1
    for index in range(n):
        if(outOfPlace >= 0):

            if((arr[index] >= 0 and arr[outOfPlace] < 0) or
            (arr[index] < 0 and arr[outOfPlace] >= 0)):
                arr = rightRotate(arr, n, outOfPlace, index)
                if(index-outOfPlace > 2):
                    outOfPlace += 2
                else:
                    outOfPlace = - 1

        if(outOfPlace == -1):

            if((arr[index] >= 0 and index % 2 == 0) or
            (arr[index] < 0 and index % 2 == 1)):
                outOfPlace = index
    return arr

# example usage:
# example 1:
arr = [1, 2, 3, -4, -1, 4]

print("from Example_1, Rearranged array is:")
print(rearrange(arr, len(arr)))

print(" ")

# example 2:
arr = [-5, -2, 5, 2, 4, 7, 1, 8, 0, -8]

print("from Example_2, Rearranged array is:")
print(rearrange(arr, len(arr)))
```

```python
# SOLUTION: 6
# The time complexity of this algorithm is O(n1 + n2), where n1 and n2 are the sizes of the two input arrays. The space complexity of this algorithm is also O(n1 + n2), since we create a new array of size n1 + n2 to store the merged elements.

# CODE:

def mergeArrays(arr1, arr2):
    n1 = len(arr1)
    n2 = len(arr2)
    arr3 = [None] * (n1 + n2)
    i = 0
    j = 0
    k = 0
    while (i < n1 and j < n2):
        if (arr1[i] < arr2[j]):
            arr3[k] = arr1[i]
            k += 1
            i += 1
        else:
            arr3[k] = arr2[j]
            k += 1
            j += 1
    while (i < n1):
        arr3[k] = arr1[i]
        k += 1
        i += 1
    while (j < n2):
        arr3[k] = arr2[j]
        k += 1
        j += 1
    return arr3

# example usage:
# example 1:
arr1 = [1, 3, 4, 5]
arr2 = [2, 4, 6, 8]
print("Output of example_1 is: ", mergeArrays(arr1, arr2))

print(" ")

# example 2:
arr1 = [5, 8, 9]
arr2 = [4, 7, 8]
print("Output of example_2 is: ", mergeArrays(arr1, arr2))
```

```python
# SOLUTION: 7
# The time complexity of this algorithm is O(n+m), where n and m are the lengths of the two input arrays. The space complexity is also O(n+m), since we use a hash set to store the elements of one array.

# CODE:

def intersection(nums1, nums2):
    set1 = set(nums1)
    result = []
    for num in nums2:
        if num in set1:
            result.append(num)
            set1.remove(num)
    return result

# example usage:
# example 1:
nums1 = [1,2,2,1]
nums2 = [2,2]
print("Output of example_1 is: ", intersection(nums1, nums2))

print(" ")

# example 2:
nums1 = [4,9,5]
nums2 = [9,4,9,8,4]
print("Output of example_2 is: ", intersection(nums1, nums2))
```

```python
# SOLUTION: 8
# The time complexity of this algorithm is O(n+m), where n and m are the lengths of the two input arrays. The space complexity is also O(n+m), since we use a hash map to store the elements of one array.

# CODE:

def intersect(nums1, nums2):
    freq = {}
    for num in nums1:
        freq[num] = freq.get(num, 0) + 1
    res = []
    for num in nums2:
        if num in freq and freq[num] > 0:
            res.append(num)
            freq[num] -= 1
    return res

# example usage:
# example 1:
nums1 = [1,2,2,1]
nums2 = [2,2]
print("Output of example_1 is: ", intersect(nums1, nums2))

print(" ")

# example 2:
nums1 = [4,9,5]
nums2 = [9,4,9,8,4]
print("Output of example_2 is: ", intersect(nums1, nums2))
```
