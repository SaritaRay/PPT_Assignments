```python
#SOLUTION: 1

#CODE:

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def maxList(l1, l2):
    if not l1:
        return l2
    if not l2:
        return l1

    if l1.data > l2.data:
        result = l1
        result.next = maxList(l1.next, l2)
    else:
        result = l2
        result.next = maxList(l1, l2.next)

    return result

# Example usage
l1 = Node(5)
l1.next = Node(2)
l1.next.next = Node(3)
l1.next.next.next = Node(8)

l2 = Node(1)
l2.next = Node(7)
l2.next.next = Node(4)
l2.next.next.next = Node(5)

result = maxList(l1, l2)

while result:
    print(result.data, end=' ')
    result = result.next

```

```python
#SOLUTION: 2
#The time complexity of this algorithm is O(n), where n is the number of nodes in the linked list. The space complexity of this algorithm is O(1), since we are not using any extra space to store the nodes.

#CODE: 

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def removeDuplicates(self):
        current = self.head

        while current is not None:
            nextNode = current.next

            while nextNode is not None and nextNode.data == current.data:
                nextNode = nextNode.next

            current.next = nextNode
            current = nextNode

    def push(self, new_data):
        new_node = Node(new_data)
        new_node.next = self.head
        self.head = new_node

    def printList(self):
        temp = self.head
        while temp:
            print(temp.data, end=" ")
            temp = temp.next

# Example 1. input:
llist = LinkedList()
llist.push(60)
llist.push(43)
llist.push(43)
llist.push(21)
llist.push(11)
llist.push(11)
llist.push(11)

print("Linked List before removing duplicates :")
llist.printList()

llist.removeDuplicates()

print("\nLinked List after removing duplicates :")
llist.printList()


# Example 2. input:
llist = LinkedList()
llist.push(34)
llist.push(25)
llist.push(25)
llist.push(25)
llist.push(12)
llist.push(12)
llist.push(10)

print("Linked List before removing duplicates :")
llist.printList()

llist.removeDuplicates()

print("\nLinked List after removing duplicates :")
llist.printList()
```

```python
#SOLUTION: 4
#The time complexity of this algorithm is O(n) where n is the number of nodes in the linked list. The space complexity is O(1) because we are only using a constant amount of extra space to store the pointers and variables.

#CODE:

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def reverse_alternate_k_nodes(self, head, k):
        if not head:
            return None

        current = head
        next_node = None
        prev = None
        count = 0

        while current and count < k:
            next_node = current.next
            current.next = prev
            prev = current
            current = next_node
            count += 1

        if head:
            head.next = current

        count = 0
        while current and count < k - 1:
            current = current.next
            count += 1

        if current:
            current.next = self.reverse_alternate_k_nodes(current.next, k)

        return prev

    def push(self, new_data):
        new_node = Node(new_data)
        new_node.next = self.head
        self.head = new_node

    def print_list(self):
        temp = self.head
        while temp:
            print(temp.data, end=" ")
            temp = temp.next


llist = LinkedList()
llist.push(9)
llist.push(8)
llist.push(7)
llist.push(6)
llist.push(5)
llist.push(4)
llist.push(3)
llist.push(2)
llist.push(1)

print("Given linked list")
llist.print_list()
llist.head = llist.reverse_alternate_k_nodes(llist.head, 3)

print("\nReversed every alternate k nodes")
llist.print_list()
```

```python
#SOLUTION: 5
#The time complexity of this algorithm is O(n) where n is the number of nodes in the linked list. The space complexity of the algorithm is O(1) because it does not use any extra space beyond a fixed number of variables.

#CODE:

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def delete_last_occurrence(self, key):
        temp = self.head
        prev = None

        while temp:
            if temp.data == key:
                last_occurrence = prev

            prev = temp
            temp = temp.next

        if last_occurrence is None:
            self.head = self.head.next
        else:
            last_occurrence.next = last_occurrence.next.next

    def push(self, new_data):
        new_node = Node(new_data)
        new_node.next = self.head
        self.head = new_node

    def print_list(self):
        temp = self.head
        while temp:
            print(temp.data, end=" ")
            temp = temp.next


llist = LinkedList()
llist.push(10)
llist.push(2)
llist.push(5)
llist.push(3)
llist.push(2)
llist.push(1)

print("Given linked list")
llist.print_list()
llist.delete_last_occurrence(2)

print("\nLinked list after deleting last occurrence of 2")
llist.print_list()
```

```python
#SOLUTION: 6
#The time complexity of the algorithm is O(m+n), where m and n are the lengths of the two linked lists. The space complexity is O(1), as we are not using any extra space to store the merged list.

#CODE:

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def mergeLists(head1, head2):
    dummyNode = Node(0)
    tail = dummyNode
    while True:
        if head1 is None:
            tail.next = head2
            break
        if head2 is None:
            tail.next = head1
            break
        if head1.data <= head2.data:
            tail.next = head1
            head1 = head1.next
        else:
            tail.next = head2
            head2 = head2.next
        tail = tail.next
    return dummyNode.next

# Define linked lists a and b
a = Node(5)
a.next = Node(10)
a.next.next = Node(15)

b = Node(2)
b.next = Node(3)
b.next.next = Node(20)

# Merge linked lists a and b
mergedListHead = mergeLists(a, b)

# Print the merged list
while mergedListHead is not None:
    print(mergedListHead.data, end=' ')
    mergedListHead = mergedListHead.next


a = Node(1)
a.next = Node(1)

b = Node(2)
b.next = Node(4)

# Merge linked lists a and b
mergedListHead = mergeLists(a, b)

# Print the merged list
while mergedListHead is not None:
    print(mergedListHead.data, end=' ')
    mergedListHead = mergedListHead.next
```

```python
#SOLUTION: 7
#The time complexity of this algorithm to reverse a doubly linked list is O(n), where n is the number of nodes in the list. This is because this function visits each node once and performs a constant amount of work on each node. The space complexity of this algorithm is O(1) because we are only using a constant amount of extra space to store pointers and temporary variables

#CODE:

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

def reverse(head):
    if head is None or head.next is None:
        return head

    current = head
    prev = None

    while current:
        next_node = current.next
        current.next = prev
        current.prev = next_node
        prev = current
        current = next_node

    return prev

head = Node(10)
head.next = Node(8)
head.next.prev = head
head.next.next = Node(4)
head.next.next.prev = head.next
head.next.next.next = Node(2)
head.next.next.next.prev = head.next.next

new_head = reverse(head)

while new_head:
    print(new_head.data, end=' ')
    new_head = new_head.next

```

```python
#SOLUTION: 8
#The time complexity of deleting a node from a doubly linked list is O(n) in the worst case where n is the number of nodes in the doubly linked list. The space complexity is O(1) as we are only deleting one node at a time and not creating any new nodes.

#CODE:

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

def deleteNode(head_ref, x):
    if head_ref is None or x <= 0:
        return head_ref

    if x == 1:
        head_ref = head_ref.next
        head_ref.prev = None
        return head_ref

    current = head_ref
    count = 1

    while current is not None and count != x:
        current = current.next
        count += 1

    if current is None:
        return head_ref

    current.prev.next = current.next

    if current.next is not None:
        current.next.prev = current.prev

    return head_ref


head = Node(1)
node2 = Node(3)
node3 = Node(4)

head.next = node2
node2.prev = head

node2.next = node3
node3.prev = node2

print("Original Doubly Linked List:")
current_node = head
while current_node is not None:
    print(current_node.data)
    current_node = current_node.next

head = deleteNode(head, 3)

print("Doubly Linked List after deleting the node at position 3:")
current_node = head
while current_node is not None:
    print(current_node.data)
    current_node = current_node.next


head = Node(1)
node2 = Node(5)
node3 = Node(2)
node4 = Node(9)

head.next = node2
node2.prev = head

node2.next = node3
node3.prev = node2

node3.next = node4
node4.prev = node3

print("Original Doubly Linked List:")
current_node = head
while current_node is not None:
    print(current_node.data)
    current_node = current_node.next

head = deleteNode(head, 1)

print("Doubly Linked List after deleting the node at position 3:")
current_node = head
while current_node is not None:
    print(current_node.data)
    current_node = current_node.next
```
