```python
#SOLUTION: 1
#The time complexity of this algorithm is O(n), where n is the number of nodes in the linked list. This is because we traverse the linked list once to find the middle node. The space complexity is O(1), because we only use a constant amount of extra space to store pointers and variables.

#CODE:

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def deleteMid(head):
    if (head == None):
        return None
    if (head.next == None):
        del head
        return None
    slow_ptr = head
    fast_ptr = head
    prev = None
    while (fast_ptr != None and fast_ptr.next != None):
        fast_ptr = fast_ptr.next.next
        prev = slow_ptr
        slow_ptr = slow_ptr.next
    prev.next = slow_ptr.next
    del slow_ptr
    return head

def printList(head):
    while (head != None):
        print(head.data, end=" ")
        head = head.next

#Taking input from example 1:
head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(4)
head.next.next.next.next = Node(5)

print("Original Linked List from Example 1: ", end="")
printList(head)

head = deleteMid(head)

print("\nModified Linked List: ", end="")
printList(head)

print(" ")
print(" ")

#Taking input from example 2:
head = Node(2)
head.next = Node(4)
head.next.next = Node(6)
head.next.next.next = Node(7)
head.next.next.next.next = Node(5)
head.next.next.next.next.next = Node(1)

print("Original Linked List from Example 2: ", end="")
printList(head)

head = deleteMid(head)

print("\nModified Linked List: ", end="")
printList(head)
```

```python
#SOLUTION: 2
#The time complexity of this Algorithm is O(n), where n is the number of nodes in the linked list. The space complexity is O(1), which means that it uses a constant amount of memory.

#CODE:

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def detectLoop(head):
    if (head == None):
        return False
    slow_ptr = head
    fast_ptr = head.next
    while (fast_ptr != None and fast_ptr.next != None):
        if (slow_ptr == fast_ptr):
            return True
        slow_ptr = slow_ptr.next
        fast_ptr = fast_ptr.next.next
    return False

# Create a linked list with the given input of example 1:
head = Node(1)
head.next = Node(3)
head.next.next = Node(4)
head.next.next.next = head.next

print("output of example 1:")
if (detectLoop(head)):
    print("True")
else:
    print("False")

print(" ")

# Create a linked list with the given input of example 2:
head = Node(1)
head.next = Node(8)
head.next.next = Node(3)
head.next.next.next = Node(4)

print("output of example 2:")
if (detectLoop(head)):
    print("True")
else:
    print("False")
```

```python
#SOLUTION: 3
#The time complexity of this algorithm is O(n), where n is the length of the linked list. The space complexity of this algorithm is O(1), which means that it uses a constant amount of memory regardless of the size of the input.

#CODE:

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
def getNthFromLast(head, N):
    first = head
    second = head
    for i in range(N):
        if second is None:
            return -1
        second = second.next
    while second is not None:
        first = first.next
        second = second.next
    return first.data

#taking input from example 1:
head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(4)
head.next.next.next.next = Node(5)
head.next.next.next.next.next = Node(6)
head.next.next.next.next.next.next = Node(7)
head.next.next.next.next.next.next.next = Node(8)
head.next.next.next.next.next.next.next.next = Node(9)

N = 2

print("output of example_1 is:", getNthFromLast(head, N))

print(" ")

#taking input from example 2:
head = Node(10)
head.next = Node(5)
head.next.next = Node(100)
head.next.next.next = Node(5)

N = 5

print("output of example_2 is:", getNthFromLast(head, N))
```

```python
#SOLUTION: 4
#The time complexity is O(n) because we traverse the linked list once to find the middle node and then traverse it again to reverse the second half of the linked list. The space complexity is O(1) because we only use a constant amount of extra space to store pointers.

#CODE:

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def isPalindrome(head):
    slow = head
    fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
    prev = None
    while slow:
        next_node = slow.next
        slow.next = prev
        prev = slow
        slow = next_node
    while prev:
        if prev.val != head.val:
            return False
        prev = prev.next
        head = head.next
    return True

#taking 1st input:
input_str = "RADAR"
head = ListNode(input_str[0])
curr_node = head
for i in range(1, len(input_str)):
    curr_node.next = ListNode(input_str[i])
    curr_node = curr_node.next

print("the output of 1st input is:")
if isPalindrome(head):
    print("Yes")
else:
    print("No")

print(" ")

#taking 2nd input:
input_str = "CODE"
head = ListNode(input_str[0])
curr_node = head
for i in range(1, len(input_str)):
    curr_node.next = ListNode(input_str[i])
    curr_node = curr_node.next

print("the output of 2nd input is:")
if isPalindrome(head):
    print("Yes")
else:
    print("No")
    
```

```python
#SOLUTION: 6
#The time complexity of this function is O(n) where n is the number of nodes in the linked list. The space complexity of this function is O(1) because we are not using any extra space to store the nodes.

#CODE:

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def skipMdeleteN(head, M, N):
    if (M == 0):
        return None

    curr = head
    prev = None

    while (curr):
        # skip M nodes
        for count in range(1, M + 1):
            if curr == None:
                return head
            prev = curr
            curr = curr.next

        # delete N nodes
        for count in range(1, N + 1):
            if curr == None:
                break
            prev.next = curr.next
            curr = prev.next

    return head

def printList(node):
    while (node != None):
        print(node.data, end=' ')
        node = node.next

#taking input from example 1:
head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(4)
head.next.next.next.next = Node(5)
head.next.next.next.next.next = Node(6)
head.next.next.next.next.next.next = Node(7)
head.next.next.next.next.next.next.next = Node(8)

M = 2
N = 2

print("EXAMPLE 1:")
print("Linked list before deletion:")
printList(head)

head = skipMdeleteN(head, M, N)

print("\nLinked list after deletion:")
printList(head)

print(" ")
print(" ")

#taking input from example2:
head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(4)
head.next.next.next.next = Node(5)
head.next.next.next.next.next = Node(6)
head.next.next.next.next.next.next = Node(7)
head.next.next.next.next.next.next.next = Node(8)
head.next.next.next.next.next.next.next.next = Node(9)
head.next.next.next.next.next.next.next.next.next = Node(10)

M = 3
N = 2

print("EXAMPLE 2:")
print("Linked list before deletion:")
printList(head)

head = skipMdeleteN(head, M, N)

print("\nLinked list after deletion:")
printList(head)

print(" ")
print(" ")

#taking input from example 3:
head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(4)
head.next.next.next.next = Node(5)
head.next.next.next.next.next = Node(6)
head.next.next.next.next.next.next = Node(7)
head.next.next.next.next.next.next.next = Node(8)
head.next.next.next.next.next.next.next.next = Node(9)
head.next.next.next.next.next.next.next.next.next = Node(10)

M = 1
N = 1

print("EXAMPLE 3:")
print("Linked list before deletion:")
printList(head)

head = skipMdeleteN(head, M, N)

print("\nLinked list after deletion:")
printList(head)
```

```python
#SOLUTION: 7
#The time complexity of this algorithm is O(min(n1, n2)) where n1 and n2 represents the length of the given two linked lists. The space complexity is O(1), no extra space is required, so it is a constant.

#CODE:

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def push(self, new_data):
        new_node = Node(new_data)
        new_node.next = self.head
        self.head = new_node

    def printList(self):
        temp = self.head
        while(temp):
            print(temp.data,end=" ")
            temp = temp.next

    def merge(self, q):
        p_curr = self.head
        q_curr = q.head

        while(p_curr != None and q_curr != None):

            p_next = p_curr.next
            q_next = q_curr.next

            q_curr.next = p_next
            p_curr.next = q_curr

            p_curr = p_next
            q_curr = q_next

        q.head = q_curr

#taking 1st input:
if __name__ == '__main__':
    llist1 = LinkedList()
    llist2 = LinkedList()
    llist1.push(11)
    llist1.push(13)
    llist1.push(17)
    llist1.push(7)
    llist1.push(5)
    
    print("FROM 1st INPUT:")

    print("First Linked List:")
    llist1.printList()

    llist2.push(6)
    llist2.push(4)
    llist2.push(2)
    llist2.push(10)
    llist2.push(12)

    print("\nSecond Linked List:")
    llist2.printList()

    llist1.merge(llist2)

    print("\nModified First Linked List:")
    llist1.printList()

    
print(" ")
print(" ")

#taking 2nd input:
if __name__ == '__main__':
    llist1 = LinkedList()
    llist2 = LinkedList()
    llist1.push(3)
    llist1.push(2)
    llist1.push(1)
    
    print("FROM 2nd INPUT:")

    print("First Linked List:")
    llist1.printList()

    llist2.push(8)
    llist2.push(7)
    llist2.push(6)
    llist2.push(5)
    llist2.push(4)

    print("\nSecond Linked List:")
    llist2.printList()

    llist1.merge(llist2)

    print("\nModified First Linked List:")
    llist1.printList()
```

```python
#SOLUTION: 8
#The time complexity of this Algorithm is O(n) and the space complexity is O(1)

#CODE:
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def isCircular(head):
    if head == None:
        return True
    slow = head
    fast = head.next
    while (fast != None and fast.next != None):
        if (slow == fast):
            return True
        slow = slow.next
        fast = fast.next.next
    return False

# Create the linked list
head = Node(10)
head.next = Node(12)
head.next.next = Node(14)
head.next.next.next = Node(16)

# Check if the linked list is circular or not
if isCircular(head):
    print("The linked list is circular.")
else:
    print("The linked list is not circular.")
```
