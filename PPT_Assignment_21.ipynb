```python
# SOLUTION: 1
# The time complexity of this algorithm is O(nlogn) and the space complexity is O(n) where n is the number of nodes in the binary tree.

# CODE:

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def inorder(root, arr):
    if not root:
        return
    inorder(root.left, arr)
    arr.append(root.val)
    inorder(root.right, arr)

def createBST(arr):
    if not arr:
        return None
    mid = len(arr) // 2
    root = TreeNode(arr[mid])
    root.left = createBST(arr[:mid])
    root.right = createBST(arr[mid+1:])
    return root

def convertBTBST(root):
    arr = []
    inorder(root, arr)
    arr.sort()
    return createBST(arr)

root = TreeNode(10)
root.left = TreeNode(2)
root.right = TreeNode(7)
root.left.left = TreeNode(8)
root.left.right = TreeNode(4)

new_root = convertBTBST(root)

def inorderTraversal(node):
  if node is None:
      return []
  left_traversal = inorderTraversal(node.left)
  right_traversal = inorderTraversal(node.right)
  return left_traversal + [node.val] + right_traversal

print(inorderTraversal(new_root))
```

```python
# SOLUTION: 2
# The time complexity of this algorithm is O(n), where n is the number of nodes in the BST. This is because we perform an inorder traversal of the tree, which visits each node exactly once. The space complexity is O(h), where h is the height of the tree.

# CODE:

class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

def findDistance(root, n1, n2):
    if root is None:
        return 0

    if root.val > n1 and root.val > n2:
        return findDistance(root.left, n1, n2)

    if root.val < n1 and root.val < n2:
        return findDistance(root.right, n1, n2)

    if root.val >= n1 and root.val <= n2 or root.val <= n1 and root.val >= n2:
        return (findLevel(root, n1, 0) + findLevel(root, n2, 0))

def findLevel(root, k, level):
    if root is None:
        return -1

    if root.val == k:
        return level

    left = findLevel(root.left, k, level + 1)
    if left == -1:
        return findLevel(root.right, k, level + 1)

    return left

# Driver code
root = Node(8)
root.left = Node(3)
root.right = Node(10)
root.left.left = Node(1)
root.left.right = Node(6)
root.left.right.left = Node(4)
root.left.right.right = Node(7)
root.right.right = Node(14)
root.right.right.left = Node(13)

n1 = 6
n2 = 14
print("The distance between",n1,"and",n2,"is",findDistance(root,n1,n2))

n3 = 3
n4 = 4
print("The distance between",n3,"and",n4,"is",findDistance(root,n3,n4))
```

```python
# SOLUTION: 3
# The time complexity of this algorithm is O(n) and the space complexity is O(n) where n is the number of nodes in the binary tree

# CODE:

class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

def inorder(root):
    if not root:
        return []
    left_traversal = inorder(root.left)
    right_traversal = inorder(root.right)
    return left_traversal + [root] + right_traversal

def convertToDLL(root):
    arr = inorder(root)
    for i in range(len(arr)-1):
        arr[i].right = arr[i+1]
        arr[i+1].left = arr[i]
    return arr[0]

root = Node(10)
root.left = Node(5)
root.right = Node(20)
root.right.left = Node(30)
root.right.right = Node(35)

head = convertToDLL(root)

while head:
    print(head.val, end=" ")
    head = head.right
```

```python
# SOLUTION: 4
# The time complexity of this algorithm is O(n) where n is the number of nodes in the binary tree. The space complexity is O(w) where w is the maximum width of the tree.

# CODE:

class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None
        self.nextRight = None

def connect(root):
    if not root:
        return root
    q = [root]
    while q:
        n = len(q)
        for i in range(n):
            node = q.pop(0)
            if i < n-1:
                node.nextRight = q[0]
            if node.left:
                q.append(node.left)
            if node.right:
                q.append(node.right)
    return root

root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
root.right.left = Node(6)
root.right.right = Node(7)

connect(root)

print(root.nextRight)
print(root.left.nextRight.val)
print(root.right.nextRight)
print(root.left.left.nextRight.val)
print(root.left.right.nextRight.val)
print(root.right.left.nextRight.val)
print(root.right.right.nextRight)
```
