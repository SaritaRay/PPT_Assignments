```python
# SOLUTION: 1
# The time complexity of this algorithm is O(n) where n is the length of the string. The space complexity is O(h) where h is the height of the tree.

# CODE:

def find_depth(tree):
    n = len(tree)
    index = [0]
    return find_depth_recursion(tree, n, index)

def find_depth_recursion(tree, n, index):
    if index[0] >= n or tree[index[0]] == 'l':
        return 0

    index[0] += 1
    left = find_depth_recursion(tree, n, index)

    index[0] += 1
    right = find_depth_recursion(tree, n, index)

    return max(left, right) + 1

# Example usage:

# Example_1:
tree = "nlnll"
print("Tree = ", tree)
print("it's Depth is = ", find_depth(tree))

print(" ")

# Example_2:
tree = "nlnnlll"
print("Tree = ", tree)
print("it's Depth is = ", find_depth(tree))
```

```python
# SOLUTION: 2
# The time complexity of this algorithm is O(n) where n is the number of nodes in the binary tree. The space complexity is O(w) where w is the maximum width of the binary tree.

# CODE:

class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

def left_view(root):
    if not root:
        return []

    queue = [root]
    left_view = []

    while queue:
        size = len(queue)

        for i in range(size):
            node = queue.pop(0)

            if i == 0:
                left_view.append(node.val)

            if node.left:
                queue.append(node.left)

            if node.right:
                queue.append(node.right)

    return left_view

# Example usage:
# Example_1:
root = Node(4)
root.left = Node(5)
root.right = Node(2)
root.right.left = Node(3)
root.right.right = Node(1)
root.right.left.left = Node(6)
root.right.left.right = Node(7)

print("the left view of the Binary Tree (given in Example_1) is = ", left_view(root))

print(" ")

# Example_2:
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.right = Node(4)
root.left.right.right = Node(5)
root.left.right.right.right = Node(6)

print("the left view of the Binary Tree (given in Example_2) is = ", left_view(root))
```

```python
# SOLUTION: 3
# The time complexity of this algorithm is O(n) where n is the number of nodes in the binary tree. The space complexity is O(w) where w is the maximum width of the binary tree.

# CODE:

class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

def right_view(root):
    if not root:
        return []

    queue = [root]
    right_view = []

    while queue:
        size = len(queue)

        for i in range(size):
            node = queue.pop(0)

            if i == size - 1:
                right_view.append(node.val)

            if node.left:
                queue.append(node.left)

            if node.right:
                queue.append(node.right)

    return right_view

# Example usage:
# Example_1:
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
root.right.left = Node(6)
root.right.right = Node(7)
root.right.right.right = Node(8)

print("Right view of the tree (given in Example_1) is: ", right_view(root))

print(" ")

# Example_2:
root = Node(1)
root.left = Node(8)
root.left.left = Node(7)

print("Right view of the tree (given in Example_2) is: ", right_view(root))
```

```python
# SOLUTION: 4
# The time complexity of this algorithm is O(n) where n is the number of nodes in the binary tree. The space complexity is O(w) where w is the maximum width of the binary tree.

# CODE:

class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

def bottom_view(root):
    if not root:
        return []

    queue = [(root, 0)]
    bottom_view = {}

    while queue:
        node, hd = queue.pop(0)

        bottom_view[hd] = node.val

        if node.left:
            queue.append((node.left, hd - 1))

        if node.right:
            queue.append((node.right, hd + 1))

    return [bottom_view[i] for i in sorted(bottom_view)]

# Example usage:
# Example_1:
root = Node(20)
root.left = Node(8)
root.right = Node(22)
root.left.left = Node(5)
root.left.right = Node(3)
root.right.right = Node(25)
root.left.right.left = Node(10)
root.left.right.right = Node(14)

print("The Bottom View from Left to Right of the given Binary Tree (in example_1) is : ", bottom_view(root))

print(" ")

# Example_2:
root = Node(20)
root.left = Node(8)
root.right = Node(22)
root.left.left = Node(5)
root.left.right = Node(3)
root.right.left = Node(4)
root.right.right = Node(25)
root.left.right.left = Node(10)
root.left.right.right = Node(14)

print("The Bottom View from Left to Right of the given Binary Tree (in example_2) is : ", bottom_view(root))
```
